```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)
```

```python
def qq_plot():
    stats.probplot(data, dist="norm", plot=plt)
    plt.title("QQ Plot")
    plt.grid(True)
    plt.show()
qq_plot()
```
