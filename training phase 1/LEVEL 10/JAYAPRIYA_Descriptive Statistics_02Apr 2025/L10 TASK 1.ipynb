```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic_dataset.csv")
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
print("Mean Values:\n", df.mean(numeric_only=True))
```

```python
print("\nMedian Values:\n", df.median(numeric_only=True))
```

```python
print("\nMode Values:\n", df.mode().iloc[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=df['Sex'], y=df['Survived'])
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x=df['Pclass'], y=df['Survived'])
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
sns.histplot(df[df['Survived'] == 1]['Age'], bins=30, kde=True, color='green', label="Survived")
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, kde=True, color='red', label="Not Survived")
plt.legend()
plt.title("Age Distribution of Survivors vs Non-Survivors")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
